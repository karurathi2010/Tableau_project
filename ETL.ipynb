```python
import pandas as pd
from datetime import datetime as dt
```

```python
bike_df=pd.read_csv("./202405-citibike-tripdata_5.csv")
```

```python
bike_df.head()
```

```python
bike_csv=bike_df.dropna()
```

```python
bike_csv.info()
```

```python
bike_csv["started_at"] = pd.to_datetime(bike_csv["started_at"])
bike_csv["ended_at"] = pd.to_datetime(bike_csv["ended_at"])
```

```python
bike_csv["start_date"] = bike_csv["started_at"].dt.date
bike_csv["start_time"] = bike_csv["started_at"].dt.time
```

```python
bike_csv["end_date"] = bike_csv["ended_at"].dt.date
bike_csv["end_time"] = bike_csv["ended_at"].dt.time
```

```python
bike_csv.head()
```

```python
del bike_csv['started_at']
del bike_csv['ended_at']
```

```python
bike_csv['start_time'] = pd.to_datetime(bike_csv['start_time'], format='%H:%M:%S')
```

```python
bike_csv['end_time'] = pd.to_datetime(bike_csv['end_time'], format='%H:%M:%S')
```

```python
bike_csv['duration']=bike_csv['end_time']-bike_csv['start_time']
```

```python
del bike_csv["duration1"]
```

```python
bike_csv['duration1']=bike_csv['duration'].dt.components.minutes
```

```python
bike_csv['duration2']=bike_csv['duration'].dt.components.seconds
```

```python
bike_csv['duration2']=bike_csv['duration2']/60
```

```python
bike_csv['total_duration']=bike_csv['duration1']+bike_csv['duration2']
```

```python
del bike_csv['duration']
del bike_csv['duration1']
del bike_csv['duration2']
```

```python
bike_csv.dropna()
```

```python
bike_csv.to_csv("./bike_data.csv")
```
